```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType,StructField, StringType, IntegerType

dataDF = [(('James','','Smith'),'1991-04-01','M',"3000"),
  (('Michael','Rose',''),'2000-05-19','M',None),
  (('Robert','','Williams'),'1978-09-05','M',"400.0"),
  (('Maria','Anne','Jones'),'1967-12-01','F',"40.00"),
  (('Jen','Mary','Brown'),'1980-02-17','F',np.nan)
]

schema = StructType([
        StructField('Name', StructType([
             StructField('firstname', StringType(), True),
             StructField('middlename', StringType(), True),
             StructField('lastname', StringType(), True)
             ])),
         StructField('DOB', StringType(), True),
         StructField('GenderMaleOrFemale', StringType(), True),
         StructField('Number', StringType(), True)
         ])

spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()
df = spark.createDataFrame(data = dataDF, schema = schema)
df.printSchema()
```

```python
df = df.fillna("0",["Number"])
df.show()
```

```python
import re
```

```python
def to_snake_case(name):
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    name = re.sub('__([A-Z])', r'_\1', name)
    name = re.sub('([a-z0-9])([A-Z])', r'\1_\2', name)
    return name.lower()
```

```python
def teste(df):
    for column in df.columns:
        df = df.withColumnRenamed(column, to_snake_case(column))
    return df
```

```python
df = teste(df)
```
